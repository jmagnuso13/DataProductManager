```python
from matplotlib.animation import FuncAnimation
%matplotlib qt5
%matplotlib inline
import numpy as np
```

```python
x_data = []
y_data = []
fig, ax = plt.subplots()
ax.set_xlim(12220,102720)
ax.set_ylim(0,175000)
line, = ax.plot(0,0)

def animation_frame(i):
    x_data.apend(i*10)
    y_data.apend(i)
    
    line.setx_data(x_data)
    line.sety_data(y_data)
    return line,

animation = FuncAnimation(fig, func=animation_frame, frames=np.arange(0, 10, 0.1), interval=10)
```
